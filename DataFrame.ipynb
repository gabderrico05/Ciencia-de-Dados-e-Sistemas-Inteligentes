```python
import pandas as pd
```

```python
# Cria um DataFrame

datas = {'nome': ['Argentina', 'Brasil', 'França', 'Itália', 'Reino Unido'],
         'continente': ['Ámerica', 'Ámerica', 'Europa', 'Europa', 'Europa'],
         'extensão': [2780400, 8515767, 643951, 301336, 243610],
         'corVerde': [0, 1, 0, 1, 0]
         }

codes = ['AR', 'BR', 'FR', 'IT', 'UK']

countries = pd.DataFrame(datas, index=codes)

print(countries)
print("--------------------------")
countries
```

```python
# Analise inicial recomendada
countries.info()
```

```python
# Propriedades básicas dos DataFrames
num_lines = countries.shape[0]
num_columns = countries.shape[1]
position = countries.index
columns = countries.columns
countries_type = type(countries)
countries_dtype = countries.dtypes
countries_idx_dtype = countries.index.dtype

print()

print(f"Número de linhas: {num_lines}\n")
print(f"Número de colunas: {num_columns}\n")
print(f'rótulos das linhas: {position} \n')
print(f'rótulos das colunas: {columns}\n')
print(f'tipo (type) {countries_type}\n')
print(f'dtypedas colunas: \n{countries_dtype}\n')
print(f"dtype dos rótulos das linhas: {countries_idx_dtype}\n")
```

```python
# Buscas no Dataframe

# Cria o DataFrame
datas = {'nome': ['Argentina', 'Brasil', 'França', 'Itália', 'Reino Unido'],
         'continente': ['Ámerica', 'Ámerica', 'Europa', 'Europa', 'Europa'],
         'extensão': [2780400, 8515767, 643951, 301336, 243610],
         'corVerde': [0, 1, 0, 1, 0]
         }

codes = ['AR', 'BR', 'FR', 'IT', 'UK']

countries = pd.DataFrame(datas, index=codes)

# Testa se um dado de linha existe
has_BR = 'BR' in countries.index
has_US = "US" in countries.index

print("existe o rótulo 'BR'? -> ", has_BR)
print(f"esxiste o rótulo 'US'? -> {has_US}\n")
print("--------------------------------------")

#Testa se um dado rótulo de coluna existe
has_greenColor = 'corVerde' in countries.columns
has_blueColor = 'corAzul' in countries.columns

print(f"existe a coluna 'corVerde'? -> {has_greenColor}")
print(f"existe a coluna 'corAzul'? -> {has_blueColor}\n")
print("--------------------------------------")

# testa se o valor faz parte de uma coluna
has_Brasil = countries['nome'].isin(['Brasil'])
print(f"existe o valor 'Brasil na coluna 'nome'? -> \n{has_Brasil}")
```

```python
# Modificando o DataFrame

print("---DataFrane original---")
print(countries)

# Inserindo o pais japão
countries.loc['JP'] = {
    'nome': 'Japão',
    'continente': 'Ásia',
    'extensão': 372,
    'corVerde': 0
    }

print('---DataFrame Modificado----')
print(countries)

# Altera a extensão do Brasil
countries.at['BR', 'extensão'] = 8512
print('DataFrame com extensão do Brasil modificada')
print(countries)


# Remove a Argentina e o Reino Unido

countries = countries.drop(['AR', 'UK'])
print('---DataFrame com Argentina e Reino Unido removidos---')
print(countries)
```
